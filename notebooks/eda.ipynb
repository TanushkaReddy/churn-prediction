```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_excel('../data/E Commerce Dataset.xlsx', sheet_name='E Comm')

# First look
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print("Data Types:\n", df.dtypes)
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.describe()
```

```python
print("Churn Distribution:\n", df['Churn'].value_counts())
```

```python
print("\nChurn %:", df['Churn'].value_counts(normalize=True) * 100)
```

```python
sns.countplot(x='Churn', data=df, palette='Set2')
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
print("Setup done!")
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

plt.figure(figsize=(10, 4))
sns.barplot(x=missing.index, y=missing.values, hue=missing.index, palette='Reds_r', legend=False)
plt.title('Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Missing Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Columns with missing values:\n", missing)
```

```python
# Fill numerical columns with median
num_cols_with_null = df.select_dtypes(include='number').columns[df.select_dtypes(include='number').isnull().any()]
df[num_cols_with_null] = df[num_cols_with_null].fillna(df[num_cols_with_null].median())

# Confirm no more missing values
print("Missing after fix:", df.isnull().sum().sum())
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='CityTier', y='Churn', data=df, hue='CityTier', palette='Set2', legend=False)
plt.title('Churn Rate by City Tier')
plt.xlabel('City Tier (1=Metro, 2=Mid, 3=Small)')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='PreferredLoginDevice', y='Churn', data=df,
            hue='PreferredLoginDevice', palette='Set1', legend=False)
plt.title('Churn Rate by Login Device')
plt.xlabel('Device')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='Churn', y='Tenure', data=df, hue='Churn', palette='Set2', legend=False)
plt.title('Tenure vs Churn (Lower tenure = higher churn risk)')
plt.xlabel('Churn (0=No, 1=Yes)')
plt.ylabel('Tenure (months)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, annot_kws={"size": 8})
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='Churn', y='OrderCount', data=df, hue='Churn', palette='Set3', legend=False)
plt.title('Order Count vs Churn')
plt.xlabel('Churn (0=No, 1=Yes)')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='SatisfactionScore', y='Churn', data=df,
            hue='SatisfactionScore', palette='RdYlGn', legend=False)
plt.title('Churn Rate by Satisfaction Score')
plt.xlabel('Satisfaction Score (1=Low, 5=High)')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
df.to_csv('../data/cleaned_data.csv', index=False)
print("Cleaned data saved!")
print("Final shape:", df.shape)
```
